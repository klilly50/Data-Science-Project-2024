```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_csv("cleaned_train_2005.csv")
df_test = pd.read_csv("cleaned_test_2005.csv")
```

```python
df_train
```

```python
df_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
model = LinearRegression()
```

```python
model_prevYearCheckouts = model.fit(df_train[['PreviousYearCheckouts']],df_train[['FirstYearCheckouts']])
preds_prevYearCheckouts = model.predict(df_test[['PreviousYearCheckouts']])
rmses_prevYeaerCheckouts = root_mean_squared_error(df_test['FirstYearCheckouts'], preds_prevYearCheckouts)
print(rmses_prevYeaerCheckouts)
```

```python
model_checkoutMonth = model.fit(df_train[['CheckoutMonth']],df_train[['FirstYearCheckouts']])
preds_checkoutMonth = model.predict(df_test[['CheckoutMonth']])
rmses_checkoutMonth = root_mean_squared_error(df_test['FirstYearCheckouts'], preds_checkoutMonth)
print(rmses_checkoutMonth)
```

```python
model_checkoutYear = model.fit(df_train[['CheckoutYear']],df_train[['FirstYearCheckouts']])
preds_checkoutYear = model.predict(df_test[['CheckoutYear']])
rmses_checkoutYear = root_mean_squared_error(df_test['FirstYearCheckouts'], preds_checkoutYear)
print(rmses_checkoutYear)
```

```python
# One-Hot Encode the categorical variables
encoded_train = pd.get_dummies(df_train, columns=['UsageClass','MaterialType','Genre'], drop_first=True)
encoded_test = pd.get_dummies(df_test, columns=['UsageClass','MaterialType','Genre'], drop_first=True)
encoded_train
```

```python
model_usageClass = model.fit(encoded_train[['UsageClass_Physical']],df_train[['FirstYearCheckouts']])
preds_usageClass = model.predict(encoded_test[['UsageClass_Physical']])
rmses_usageClass = root_mean_squared_error(encoded_test['FirstYearCheckouts'], preds_usageClass)
print(rmses_usageClass)
```

```python
model_materialType = model.fit(encoded_train[['MaterialType_BOOK', 'MaterialType_EBOOK', 'MaterialType_OTHER', 'MaterialType_SOUNDDISC', 'MaterialType_VIDEODISC']],df_train[['FirstYearCheckouts']])
preds_materialType = model.predict(encoded_test[['MaterialType_BOOK', 'MaterialType_EBOOK', 'MaterialType_OTHER', 'MaterialType_SOUNDDISC', 'MaterialType_VIDEODISC']])
rmses_materialType = root_mean_squared_error(encoded_test['FirstYearCheckouts'], preds_materialType)
print(rmses_materialType)
```

```python
model_genre = model.fit(encoded_train[['Genre_fantasy/sci-fi', 'Genre_fiction', 'Genre_history', 'Genre_horror/thriller', 'Genre_juvenile', 'Genre_mystery', 'Genre_nonfiction', 'Genre_other', 'Genre_romance', 'Genre_young adult']],df_train[['FirstYearCheckouts']])
preds_genre = model.predict(encoded_test[['Genre_fantasy/sci-fi', 'Genre_fiction', 'Genre_history', 'Genre_horror/thriller', 'Genre_juvenile', 'Genre_mystery', 'Genre_nonfiction', 'Genre_other', 'Genre_romance', 'Genre_young adult']])
rmses_genre = root_mean_squared_error(encoded_test['FirstYearCheckouts'], preds_genre)
print(rmses_genre)
```
