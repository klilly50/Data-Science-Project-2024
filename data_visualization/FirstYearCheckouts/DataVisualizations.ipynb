```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_csv('../../data/FirstYearCheckouts/cleaned_train_full_2005.csv')
```

```python
df_train
```

```python
sns.scatterplot(data=df_train,x='PreviousYearCheckoutsSum',y='FirstYearCheckouts',hue='MaterialType')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Number of Checkouts from Author in Previous Year")
plt.ylabel("Total Checkouts in First Year")
plt.legend(title='Material Type',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.scatterplot(data=df_train,x='PreviousYearCheckouts',y='FirstYearCheckouts',hue='Genre')
plt.xlabel("Number of Checkouts from Author in Previous Year")
plt.ylabel("Total Checkouts in First Year")
plt.legend(title='Genre',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.barplot(data=df_train, x='MaterialType', y='FirstYearCheckouts',estimator=np.sum)
plt.title("Sum of Checkouts by Material Type")
plt.xlabel("Material Type")
plt.ylabel("First Year Checkouts")
plt.xticks(rotation=45)
plt.show()
```

```python
# Set up bins for the x-axis data
num_bins = 10  # Adjust this to your preference
df_train['x_bins'] = pd.cut(df_train['PreviousYearCheckouts'], bins=num_bins)

# Group by bins and calculate the sum of y-values in each bin
binned_data = df_train.groupby('x_bins')['FirstYearCheckouts'].sum().reset_index()
binned_data
```

```python
sns.barplot(data=binned_data, x='x_bins', y='FirstYearCheckouts')
plt.xlabel('Previous Year Checkouts Bins')
plt.ylabel('Sum of Y Values')
plt.title('Sum of Y within Each X Bin')
plt.xticks(rotation=45)  # Rotate x labels if needed
plt.show()
```

```python
binned_data_NoFirstBin = binned_data.iloc[1:]
binned_data_NoFirstBin
```
