```python
#use pandas to import dataset
import pandas as pd
#read in data skipping the lines that aren't formatted correctly
df = pd.read_csv('../Downloads/Checkouts_by_Title.csv')
print (df)
```

```python

from matplotlib import pyplot as plt

#plot page length versus rating. is it linear or compoletely uncorrelated?
#matplotlib to plot



fig, axs =plt.subplots(1,2)
axs[0].plot(df['average_rating'])
axs[0].set_xlabel('The number it is in the excel file i think')
axs[0].set_ylabel('Average Rating')
axs[0].set_title('Average Rating')
axs[1].plot(df['num_pages'])
axs[1].set_xlabel('The number it is in the excel file i think')
axs[1].set_ylabel('Number of Pages')
axs[1].set_title('Number of Pages')



plt.tight_layout()

df.plot.scatter(x='average_rating', y='num_pages')
plt.title("Average Rating Vs Number of Pages")
plt.xlabel("Average Rating")
plt.ylabel("Number of Pages")

df.plot(x='average_rating', y=['num_pages', 'ratings_count'])
plt.title("Average Rating Vs Number of Pages and Number of Ratings")
plt.xlabel("Average Rating")
```
