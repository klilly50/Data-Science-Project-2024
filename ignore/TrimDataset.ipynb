```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

This reads in the CSV, dropping some of the columns we don't need and turning the checkout year and checkout month into a date.

```python
df_trimmed=pd.read_csv("data/Checkouts_by_Title.csv",usecols=['MaterialType','CheckoutYear','CheckoutMonth','Checkouts','Title','Creator','Subjects','PublicationYear'], parse_dates={'dates':['CheckoutYear','CheckoutMonth']} )
```

This trims the rows that are lacking an entry.

```python
df=df_trimmed.dropna()
```

```python
df.head()
```

```python
df.info
```

To make this dataframe more manageable, we filter out all the information from 2017 and earlier.

```python
df.sort_values(by='dates')
```

```python
df_after2017=df[df['dates']>'2017-12-31']
```

```python
df_after2017.head(100)
```

```python
df_after2017.info
```

```python
df_after2017.sort_values(by=['dates'])
```

Now, we filter to only get books and ebooks.

```python
df_after2017_books=df_after2017[df_after2017['MaterialType'].isin(['BOOK','EBOOK'])]
```

```python
df_after2017_books.info
```

This is still pretty big, it has around 10 million rows, while our original dataset had 45 million rows. Let's see how many are just books.

```python
df_after2017_books[df['MaterialType']=='BOOK'].shape[0]
```

So, there are about 6 million rows in this range that are books. This is still pretty big, but a lot smaller than what we started with.

```python
df_after2017_onlyprintbooks=df_after2017_books[df_after2017_books['MaterialType']=='BOOK']
```

```python
df_after2017_onlyprintbooks.info
```

```python
df_after2017_onlyprintbooks.sort_values(by='dates')
```

```python
df_after2017_onlyprintbooks.to_csv('printbooksafter2017.csv')
```
