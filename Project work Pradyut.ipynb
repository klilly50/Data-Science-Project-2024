```python
plt.plot(df['Unnamed: 0'], df['Year-Month'], label='label for Year-Month', color='blue')
plt.title('Line plot of Unnamed:0 vs Year-Month')
plt.xlabel('Unnamed: 0')
plt.ylabel('Year-Month')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Year-Month'], df['Unnamed: 0'], color='green', s=100)
plt.title('Year-Month Plot vs Checkouts')
plt.ylabel('Checkouts')
plt.xlabel('Year-Month')
plt.show()
```

```python
for i in df.columns:
    plt.figure()
    plt.hist(df[i])
```

```python
columns=df.columns
plt.figure(figsize=(20, 5))
```

```python
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        x_col=columns[i]
        y_col=columns[j]
        plt.subplot(len(columns)-1, len(columns)-1, i* (len(columns)-1)+j)
        plt.scatter(df[x_col], df[y_col])
        plt.title(f'Scatter Plot: {x_col} vs {y_col}')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.subplots_adjust(wspace=0.4, hspace=0.4)
        plt.tight_layout()
        plt.show
```
